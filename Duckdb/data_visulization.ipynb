```python
import duckdb
import leafmap
import lonboard
```

```python
url = 'https://open.gishub.org/data/duckdb/cities.parquet'
```

```python
gdf = leafmap.read_parquet(url, return_type='gdf', src_crs='EPSG:4326')
gdf.head()
```

```python
gdf.explore()
```

```python
leafmap.view_vector(gdf, get_radius=20000, get_fill_color='blue')
```
